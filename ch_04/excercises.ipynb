Using the faang.csv file, group by the ticker and resample to monthly frequency. Make the following aggregations:
1) Mean of the opening price
2) Maximum of the high price
3) Minimum of the low price
4) Mean of the closing price
5) Sum of the volume traded

```python
faang = pd.read_csv("exercises/faang.csv", parse_dates=["date"], index_col="date")
faang.head(3)
```

```python
faang.index
```

```python
faang.groupby("ticker").resample("M").agg(
    {
        "open": np.mean,
        "high": np.max,
        "low": np.min,
        "close": np.mean,
        "volume": np.sum
        }
    )
```

Use the transform() method on the FAANG data to represent all the values in terms of the first date in the data. To do so, divide all the values for each ticker by the values for the first date in the data for that ticker. This is referred to as an index, and the data for the first date is the base (https://ec.europa.eu/eurostat/statistics-explained/index.php/Beginners:Statistical_concept_-_Index_and_base_year). When data is in this format, we can easily see growth over time. Hint: transform() can take a function name.

```python
faang = faang.reset_index().set_index(['ticker', 'date'])
```

```python
faang_index = faang.div(faang.groupby(level='ticker').transform('first'))
faang_index.groupby(level='ticker').agg('head', 3)
```

```python
covid = pd.read_csv('exercises/covid19_cases.csv')\
    .assign(date=lambda x: pd.to_datetime(x.dateRep, format='%d/%m/%Y'))\
    .set_index('date')\
    .replace('United_States_of_America', 'USA')\
    .replace('United_Kingdom', 'UK')\
    .sort_index()
```

```python
covid.index.unique()
```

```python
top_five_countries = covid\
    .groupby('countriesAndTerritories').cases.sum()\
    .nlargest(5).index
covid[covid['countriesAndTerritories'].isin(top_five_countries)].groupby('countriesAndTerritories').cases.idxmax()
```

```python
covid.groupby([pd.Grouper(freq='1D'), 'countriesAndTerritories']).cases.sum().unstack()
```

```python
covid.groupby(['countriesAndTerritories', pd.Grouper(freq='1D')]).cases.sum().unstack(0).diff().rolling(7).mean().last('1W')[top_five_countries]
```

```python
covid.head(3)
```

```python
covid\
    .pivot(columns='countriesAndTerritories', values='cases')\
    .drop(columns='China')\
    .apply(lambda x: x[x > 0].index.min())\
    .sort_values()\
    .rename(lambda x: x.replace('_', ' ')).to_frame('first date')
```

```python
covid\
    .pivot_table(columns='countriesAndTerritories', values='cases', aggfunc='sum')\
    .T\
    .transform('rank', method='max', pct=True)\
    .sort_values('cases', ascending=False)\
    .rename(lambda x: x.replace('_', ' '))
```
